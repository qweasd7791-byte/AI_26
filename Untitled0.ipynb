```python
import numpy as np
arr_1 = np.array([1,2,3,4,5]) # 파이썬 리스트를 넘파이 배열로 변환
print(arr_1) # 넘파이 배열 출력
# 파이썬 리스트는 [1,2,3,4,5]로 출력, 넘파이 배열은 [1 2 3 4 5]로 출력 -> 쉼표 유무 차이
```

```python
[1,2,3,4,5] * 2    # 파이썬 리스트에 * 2
```

```python
arr_1 * 2 # 넘파이 1차원 배열에 * 2
```

```python
# 넘파이의 데이터타입 살펴보기
import numpy as np

arr_1 = np.array([1,2,3,4,5])
print(arr_1.dtype) # 정수로 구성된 넘파일 배열의 데이터타입(기본)
```

```python
print(np.array([1,2,3], dtype='int8').dtype)
# dtype을 int8로 지정함
# dtype을 특정 데이터 타입으로 지정하여 원하는 데이터 타입으로 넘파이 배열을 만들 수 있음
```

```python
print(np.array([1,2,3], dtype='i1').dtype)
# np.array([1,2,3], dtype='int8').dtype 코드와 결과가 같음
# dtype='i1': 1 바이트 정수형 데이터 타입을 의미함
```

```python
# 넘파이 배열의 장점
# 1. 메모리 사용 효율이 좋고, 게산 속도가 빠름
# 2. 선형 대수 연산 지원
# 3. 백터화 연산 지원
```

```python
import numpy as np
ndarray_1 = np.array([1,2,3]) # 넘파이 배열
ndarray_2 = np.array([4,5,6]) # 넘파이 배열
list_1 = [1,2,3] # 파이썬 리스트
list_2 = [4,5,6] # 파이썬 리스트
```

```python
print(ndarray_1 + ndarray_2) # 넘파이 배열 연산
print(list_1 + list_2)       # 파이썬 리스트 연산
```

```python
result_list = [x +y for x, y in zip(list_1, list_2)]
print(result_list)
```

```python
print(ndarray_1 * 2) # 넘파이 배열 연산
print(list_1 * 2)    # 파이썬 리스트 연산
```

```python
# 파이썬 리스트로 넘파이 배열 생성하기
import numpy as np


a = np.array([2,3,4]) # 넘파이 정수형 배열 생성
print(a) # 넘파이 배열 출력
print(a.dtype) # 넘파이 배열의 데이터 타입의 출력
b = np.array([2.1, 3.2, 4.3]) # 넘파이 실수형 배열 생성
print(b)
print(b.dtype)
```
